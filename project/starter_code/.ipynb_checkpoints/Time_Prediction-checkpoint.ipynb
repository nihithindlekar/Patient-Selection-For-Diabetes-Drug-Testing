```python
# Re-usable boxplot function
def boxplot_los_groupby(variable, time_range=(0, 20), size=(8,4)):
    ''' 
    Boxplot of time in hospital by df categorical series name 
    '''
    results = analysis_df[[variable, 'time_in_hospital']].groupby(variable).median().reset_index()

    categories = results[variable].values.tolist()

    hist_data = []
    for cat in categories:
        hist_data.append(analysis_df['time_in_hospital'].loc[analysis_df[variable]==cat].values)

    fig, ax = plt.subplots(figsize=size)
    ax.boxplot(hist_data, 0, '', vert=False)
    ax.set_xlim(time_range)
    ax.set_yticklabels(categories)
    ax.set_xlabel('Time in Hospital (days)')
    ax.tick_params(left=False, right=False) 
    ax.set_title('Comparison of {} categories'.format(variable))
    plt.tight_layout()
    plt.show();
```

### ICD-9 Code Feature Engineering

From Wiki: International Classification of Diseases, Clinical Modification (ICD-9-CM) is an adaption created by the U.S. National Center for Health Statistics (NCHS) and used in assigning diagnostic and procedure codes associated with inpatient, outpatient, and physician office utilization in the United States.

```python
print("There are {} number of unique primary diagnosis codes".format(ndc_reduce_dim_df['primary_diagnosis_code'].nunique()))
```

```python
print("There are {} number of unique other diagnosis codes".format(ndc_reduce_dim_df['other_diagnosis_codes'].nunique()))
```

###### Because it's not feasible to have 716 unique values to use as features for predicting time in hospital, we need to reduce the diagnosis into more general categories. After researching the ICD9 approach, I discovered that they arranged into super categories as the following (source):

001–139: infectious and parasitic diseases <br>
140–239: neoplasms<br>
240–279: endocrine, nutritional and metabolic diseases, and immunity disorders<br>
280–289: diseases of the blood and blood-forming organs<br>
290–319: mental disorders<br>
320–389: diseases of the nervous system and sense organs<br>
390–459: diseases of the circulatory system<br>
460–519: diseases of the respiratory system<br>
520–579: diseases of the digestive system<br>
580–629: diseases of the genitourinary system<br>
630–679: complications of pregnancy, childbirth, and the puerperium<br>
680–709: diseases of the skin and subcutaneous tissue<br>
710–739: diseases of the musculoskeletal system and connective tissue<br>
740–759: congenital anomalies<br>
760–779: certain conditions originating in the perinatal period<br>
780–799: symptoms, signs, and ill-defined conditions<br>
800–999: injury and poisoning<br>
E and V codes: external causes of injury and supplemental classification, using 999 as placeholder even though it overlaps with complications of medical care<br>

Now the task is to re-code the primary codes in terms of these category groups.

```python

# ICD-9 Main Category ranges
icd9_ranges = [(1, 140), (140, 240), (240, 280), (280, 290), (290, 320), (320, 390), 
               (390, 460), (460, 520), (520, 580), (580, 630), (630, 680), (680, 710),
               (710, 740), (740, 760), (760, 780), (780, 800), (800, 1000), (1000, 2000)]

# Associated category names
diag_dict = {0: 'Infectious', 1: 'Neoplasms', 2: 'Diabetes', 3: 'blood',
             4: 'mental', 5: 'Nervous', 6: 'Circulatory', 7: 'Respiratory',
             8: 'Digestive', 9: 'Genitourinary', 10: 'Pregnancy', 11: 'Skin', 
             12: 'Muscular', 13: 'Congenital', 14: 'Perinatal', 15: 'Misc_Symptoms',
             16: 'Injury', 17: 'Misc_Injuries'}

# Filter out E and V codes since processing will be done on the numeric first 3 values
#create a function for reducing the ICD9 dimensions
def reduce_dimension_diagnosis_codes(df, column):
    df['recode'] = df[column]
    df['recode'] = df['recode'][~df['recode'].str.contains("[a-zA-Z]").fillna(False)]
    df['recode'].fillna(value='999', inplace=True)
    df['recode'] = df['recode'].str.slice(start=0, stop=3, step=1)
    df['recode'] = df['recode'].astype(int)
    
    # Re-code primary in terms of integer
    for num, cat_range in enumerate(icd9_ranges):
        df['recode'] = np.where(df['recode'].between(cat_range[0],cat_range[1]), 
                num, df['recode'])
    
    # Convert integer to category name using diag_dict
    df['recode'] = df['recode']
    df[column + '_' + 'cat'] = df['recode'].replace(diag_dict)
    df.drop(['recode'], axis=1, inplace=True)
    return df
```

```python
#calling the function to reduce primary code dimensions
ndc_reduce_dim_df = reduce_dimension_diagnosis_codes(ndc_reduce_dim_df, 'primary_diagnosis_code')
ndc_reduce_dim_df.head()
```

## Split Dataset into Train, Validation, and Test Partitions

**Question 6**: In order to prepare the data for being trained and evaluated by a deep learning model, we will split the dataset into three partitions, with the validation partition used for optimizing the model hyperparameters during training. One of the key parts is that we need to be sure that the data does not accidently leak across partitions.

Please complete the function below to split the input dataset into three partitions(train, validation, test) with the following requirements.
- Approximately 60%/20%/20%  train/validation/test split
- Randomly sample different patients into each data partition
- **IMPORTANT** Make sure that a patient's data is not in more than one partition, so that we can avoid possible data leakage.
- Make sure that the total number of unique patients across the splits is equal to the total number of unique patients in the original dataset
- Total number of rows in original dataset = sum of rows across all three dataset partitions

```python
def patient_dataset_splitter(df, patient_key='patient_nbr'):
    '''
    df: pandas dataframe, input dataset that will be split
    patient_key: string, column that is the patient id

    return:
     - train: pandas dataframe,
     - validation: pandas dataframe,
     - test: pandas dataframe,
    '''
    
    train, validation, test = np.split(df.sample(frac=1, random_state=1), [int(.6*len(df)), int(.8*len(df))])
    
    return train, validation, test
```

```python
#from student_utils import patient_dataset_splitter
d_train, d_val, d_test = patient_dataset_splitter(agg_encounter_df, 'patient_nbr')
```

# Baseline Model Selection

We would begin by fitting few linear and non-linear models on the base data for selecting the best model forn further tuning and improvement

```python
from sklearn.linear_model import Lasso, LinearRegression, PoissonRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from math import sqrt
```

```python
# Regression models for comparison
models = [GradientBoostingRegressor(random_state = 0), 
          LinearRegression(),
          PoissonRegressor(),
          KNeighborsRegressor(),
          RandomForestRegressor(random_state = 0),
          Lasso(random_state = 0)]

results = {}
r2score = {}

for model in models:
    
    # Instantiate and fit Regressor Model
    reg_model = model
    reg_model.fit(X_train, y_train)
    
    # Make predictions with model
    y_test_preds = reg_model.predict(X_test)
    
    # Grab model name and store results associated with model
    name = str(model).split("(")[0]
    results[name] = sqrt(mean_squared_error(y_test, y_test_preds))
    r2score[name] = r2_score(y_test, y_test_preds)
    


    print('{} done.'.format(name))
```

```python
# RMSE score results
fig, ax = plt.subplots()
ind = range(len(results))
ax.barh(ind, list(results.values()), align='center', 
        color = '#55a868', alpha=0.8)
ax.set_yticks(ind)
ax.set_yticklabels(results.keys())
ax.set_xlabel('RMSE')
ax.tick_params(left=False, top=False, right=False) 
ax.set_title('Comparison of Regression Models')
#fig.savefig('images/compare_models.png', bbox_inches = 'tight')
```

## POST Feature Selection

```python
# Regression models for comparison
models = [GradientBoostingRegressor(random_state = 0), 
          LinearRegression(),
          KNeighborsRegressor(),
          RandomForestRegressor(random_state = 0),
          ]

results = {}
adj_r2 = {}
for model in models:
    
    # Instantiate and fit Regressor Model
    reg_model = model
    reg_model.fit(X_train_selected, y_train)
    
    # Make predictions with model
    y_test_preds = reg_model.predict(X_test_selected)
    
    # Grab model name and store results associated with model
    name = str(model).split("(")[0]

    results[name] = r2_score(y_test, y_test_preds)
    r_squared = r2_score(y_test, y_test_preds)
    adjusted_r_squared = 1 - ((1-r_squared)*(len(y_train)-1))/(len(y_train)-X_train_selected.shape[1]-1)
    adj_r2[name] = adjusted_r_squared
    print('{} done.'.format(name))
```

```python
# R2 score results
fig, ax = plt.subplots()
ind = range(len(results))
ax.barh(ind, list(results.values()), align='center', 
        color = '#55a868', alpha=0.8)
ax.set_yticks(ind)
ax.set_yticklabels(results.keys())
ax.set_xlabel('R-squared score')
ax.tick_params(left=False, top=False, right=False) 
ax.set_title('Comparison of Regression Models')
#fig.savefig('images/compare_models.png', bbox_inches = 'tight')
```

```python
# R2 score results
fig, ax = plt.subplots()
ind = range(len(adj_r2))
ax.barh(ind, list(adj_r2.values()), align='center', 
        color = '#55a868', alpha=0.8)
ax.set_yticks(ind)
ax.set_yticklabels(adj_r2.keys())
ax.set_xlabel('Adjusted R-squared score')
ax.tick_params(left=False, top=False, right=False) 
ax.set_title('Comparison of Regression Models')
#fig.savefig('images/compare_models.png', bbox_inches = 'tight')
```
